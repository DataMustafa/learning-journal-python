```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis
```

```python
Premium = pd.read_csv("Premiums.csv")
```

```python
mode_policies = Premium['ZONE_NAME'].mode()[0]
print(f"Mode of policies by Zone: {mode_policies}")
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=Premium['Vintage_Period'])
plt.title("Boxplot of Vintage Period")
plt.show()
```

```python
Q1 = Premium['Vintage_Period'].quantile(0.25)
Q3 = Premium['Vintage_Period'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = Premium[(Premium['Vintage_Period'] < lower_bound) | (Premium['Vintage_Period'] > upper_bound)]
print(f"Detected Outliers:\n{outliers}")
```

```python
print("\nSkewness and Kurtosis of Premium by Zone:")
for zone in Premium['ZONE_NAME'].unique():
    zone_data = Premium[Premium['ZONE_NAME'] == zone]['Premium']
    print(f"{zone}: Skewness = {skew(zone_data.dropna())}, Kurtosis = {kurtosis(zone_data.dropna())}")
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=Premium['Vintage_Period'], y=Premium['Premium'])
plt.xlabel("Vintage Period")
plt.ylabel("Premium")
plt.title("Scatter Plot: Premium vs. Vintage Period")
plt.show()
```

```python
correlation = Premium[['Premium', 'Vintage_Period']].corr().iloc[0, 1]
print(f"Correlation coefficient between Premium and Vintage Period: {correlation}")
```

```python
if correlation > 0:
    print("Positive correlation: As Vintage Period increases, Premium also tends to increase.")
elif correlation < 0:
    print("Negative correlation: As Vintage Period increases, Premium tends to decrease.")
else:
    print("No correlation: Premium and Vintage Period are independent.")
```
